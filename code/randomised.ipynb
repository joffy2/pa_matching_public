```r
### compeltely Randomised (equal probability of all parameters)

n = 1000


set.seed(123) 

set_ups <- data.frame(method = c(rep('PSM', n*0.5), rep('MDM',n*0.5)), 
                      caliper = sample(c(0.25,0,0.5), size = n, replace = TRUE), 
                      replacement = sample(c('TRUE','FALSE'), size = n, replace = TRUE),
                      no.conf = sample(x = seq(3,10), size = n, replace = TRUE),
                      exact = sample(c('yes', 'no'), size = n, replace = TRUE))  %>%
                      mutate(distance = ifelse(method == 'MDM', 'mahalanobis', 'glm'))
                        
```

```r
##select a weighted random selection of confounders
formulas = list()
calipers = list()
exact_confs = list()
all_confs = list()

##set n + 1 as a vlaue to ensure that lists are correct length even if nubmer 500 is a NULL value
calipers[[n+1]] = n 
exact_confs [[n+1]] = n 
all_confs [[n+1]] = n 

set.seed(123) 

for (i in 1:n) {

    row = set_ups[i,]
    
    ##take a random sample of confounders
    conf_ls = sample(confounders$var_name, size=row$no.conf)
    all_confs[[i]] = conf_ls
    
        if(row$exact == 'yes' & any(str_detect(conf_ls,'veg_num|prov_num|ecoregion_num|agri_suit'))){
            exact_confs[i] <- list(stri_remove_empty(str_extract(conf_ls,'veg_num|prov_num|ecoregion_num|agri_suit'), na_empty =TRUE))
            conf_ls = stri_remove_empty(str_remove(conf_ls,'veg_num|prov_num|ecoregion_num|agri_suit'))
        }else {
            exact_confs[i] <- NULL
            }

        if(row$caliper != 0 & row$method == 'PSM'){
            calipers[i] = row$caliper 
            
        }else{
            
            if(row$caliper != 0 & row$method == 'MDM'){
                conf_ls = stri_remove_empty(str_remove(conf_ls,'veg_num|prov_num|ecoregion_num|agri_suit')) ##Mal dist caliper can't have factors in it 
                calipers[i] = list(rep(row$caliper,length(conf_ls))) ## each conf needs to be in named lsit for Mal caliper 
                names(calipers[[i]]) = conf_ls
            }
           }
    
    formulas[[i]] = formula(paste('treatment ~ ', paste(conf_ls, collapse = ' + ')))

    }
```

```r
matches = list()

for (i in 1:n) {
    tryCatch({
        set_up <- set_ups[i,]
            
        match <- matchit(formula = formulas[[i]], data = data, 
                     method = "nearest", distance = set_up$distance, 
                     replace = as.logical(set_up$replacement), 
                     caliper =  calipers[[i]],
                     exact = exact_confs[[i]],
                     ratio = 1)
        matches[[i]] <- match 
      },error=function(e){})
    }
    
```

```r
##list of matches
num <- str_pad(1:n, pad = 0,width = 3, "left")
```

```r
##get the quaility metrics of the matches 

smds = data.frame()
ptms = data.frame()

for(i in 1:n) {
    tryCatch({
        
    smd <- bal.tab(matches[[i]])$Balance %>% 
            rownames_to_column('confounder') %>% 
            rename('smd' = 'Diff.Adj') %>%
            select(confounder, smd) %>% 
            mutate(match = paste0('match_',num[i]), method = set_ups[i,1])
        
    smds <- bind_rows(smds, smd)
    
    ptm <- data.frame(percent_trt_matched = ((summary(matches[[i]], standerize = TRUE)$nn)[4,2] / 331) *100, 
                        match = paste0('match_',num[i]))    
    ptms <- bind_rows(ptms, ptm)
        },error=function(e){})
    }
```

```r
ggplot() +
       geom_point(data = smds, aes(x = smd, confounder, colour = method)) +
        geom_vline(xintercept = 0.25) +
        geom_vline(xintercept = -0.25) +
        geom_vline(xintercept = -0.1, linetype = 2) +
        geom_vline(xintercept = 0.1, linetype = 2)
```

```r
ggplot() +
       geom_histogram(data = smds, aes(x = smd, fill = method)) +
        geom_vline(xintercept = 0.25) +
        geom_vline(xintercept = -0.25) +
        geom_vline(xintercept = -0.1, linetype = 2) +
        geom_vline(xintercept = 0.1, linetype = 2)
```

```r
all_confs_df = data.frame()

for( i in 1:n) {
    
    df <- confounders %>% 
                select(var_name) %>% 
                left_join(data.frame(var_name =  all_confs[[i]], x = 1), by = 'var_name') %>% 
                    pivot_wider(names_from = var_name, values_from = x) %>% 
                        mutate(match = paste0('match_',num[i])) %>%
                        replace(is.na(.), 0)
    
    all_confs_df = bind_rows(all_confs_df, df)
    }

```

```r
##save the matching set ups for each match
match_info <- set_ups %>% 
            mutate(match = paste0('match_',num)) %>%
            left_join(all_confs_df, by = 'match') %>%
            left_join(ptms, by = 'match') %>%
            rowwise() %>%
            mutate(fac_confs = sum(veg_num, prov_num, ecoregion_num, agri_suit)) %>%
            ungroup() %>%
            mutate(exact = ifelse(exact == 'yes' & fac_confs == 0, 'no', exact))

```

```r
summary(match_info$percent_trt_matched[match_info$method == 'PSM'])
summary(match_info$percent_trt_matched[match_info$method == 'MDM'])
```
