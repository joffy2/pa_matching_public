```r
##Notebook to conducting matching for PA Mathcing study
## Create matched datasets based on commonly used methods taken from reivew ##
```

```r
library(tidyverse) 
library(MatchIt)
library(cem)
library(cowplot)
library(optmatch)
library(MetBrewer)
library(ecodist)
```

```r
##read in unmatched dataset
data1 <- read_csv('data/unmatched_data_clusters.csv') 
```

```r
##list of confounders for first analysis - the top 6
conf <- c('slope', 'elevation', 'precipitation', 'nearest_road', 'travel_time_city_2015', 'treecover_2014')
formula = formula(treatment ~ slope + elevation + precipitation + nearest_road + travel_time_city_2015 + treecover_2014)
```

```r
##remove missing values (n =2) and set cluster ID as rowname for matching
data <- data1 %>% 
            drop_na() %>%
            select(plot_cluster, treatment, all_of(conf)) %>%
                       column_to_rownames(var = "plot_cluster")
```

```r
##standardise all covarites
data_stand <- data1 %>% 
                drop_na() %>%
                    select(plot_cluster, treatment, all_of(conf)) %>%
                    mutate(across(all_of(conf), ~ (.x - mean(.x, na.rm = TRUE)) / sd(.x, na.rm = TRUE)))
```

```r
table(data$treatment)
```

```r
## method 1 - PSM 
PSM_match <- matchit(formula = formula, data = data, method = "nearest", distance = "glm", 
                     link = 'logit', replace=FALSE, ratio = 1, caliper=0.25)
```

```r
##save SMD inforamtion
PSM_smd <- as.data.frame(summary(PSM_match, standardize = TRUE)$sum.matched) %>% 
            rename('sd_mean_dif' = 'Std. Mean Diff.', 'var_ratio' = 'Var. Ratio') %>%
            mutate(conf = rownames(.), dataset = 'PSM_clpr', abs_smd = abs(sd_mean_dif))
```

```r
## method 2 - MDM1 
MDM_match <- matchit(formula = formula, data = data, method = "nearest", distance = "mahalanobis", replace=FALSE, ratio = 1, include.obj = TRUE,
                      caliper=c(slope = 0.5, elevation = 0.5, precipitation = 0.5, nearest_road = 0.5, travel_time_city_2015 = 0.5, treecover_2014 = 0.5))
```

```r
##save SMD inforamtion
MDM_smd <-as.data.frame(summary(MDM_match, standardize = TRUE)$sum.matched) %>% 
            rename('sd_mean_dif' = 'Std. Mean Diff.' , 'var_ratio' = 'Var. Ratio') %>%
            mutate(conf = rownames(.), dataset = 'MDM_clpr', abs_smd = abs(sd_mean_dif))
```

```r
## method 3 - PSM no caliper
PSM_match2 <- matchit(formula = formula, data = data, method = "nearest", distance = "glm", 
                     link = 'logit', replace=FALSE, ratio = 1)
```

```r
##save SMD inforamtion
PSM_smd2 <- as.data.frame(summary(PSM_match2, standardize = TRUE)$sum.matched) %>% 
            rename('sd_mean_dif' = 'Std. Mean Diff.', 'var_ratio' = 'Var. Ratio') %>%
            mutate(conf = rownames(.), dataset = 'PSM', abs_smd = abs(sd_mean_dif))
```

```r
## method 4 - MDM no caliper
MDM_match2 <- matchit(formula = formula, data = data, method = "nearest", distance = "mahalanobis", replace=FALSE, ratio = 1, include.obj = TRUE)
```

```r
##save SMD inforamtion
MDM_smd2 <- as.data.frame(summary(MDM_match2, standardize = TRUE)$sum.matched) %>% 
            rename('sd_mean_dif' = 'Std. Mean Diff.', 'var_ratio' = 'Var. Ratio') %>% 
            mutate(conf = rownames(.), dataset = 'MDM', abs_smd = abs(sd_mean_dif))
```

```r
##SMD for  unmatched dataset
unmatched <- as.data.frame((summary(PSM_match, standardize = TRUE)$sum.all)) %>% 
            rename('sd_mean_dif' = 'Std. Mean Diff.', 'var_ratio' = 'Var. Ratio') %>%
            mutate(dataset = 'Unmatched', conf = rownames(.), abs_smd = abs(sd_mean_dif))
```

```r
smd <- bind_rows(PSM_smd, MDM_smd, PSM_smd2, MDM_smd2, unmatched) 
```

```r
## save the datasets - with distnace scores and matched pairs
PSM_data <- match.data(PSM_match, data = data, distance = "pscore") %>% 
                   rownames_to_column(var = 'plot_cluster') %>% 
                   mutate(dataset = 'PSM_clpr')

MDM_data <- match.data(MDM_match, data = data) %>% 
              rownames_to_column(var = 'plot_cluster')  %>% 
                   mutate(dataset = 'MDM_clpr')

PSM_data2 <- match.data(PSM_match2, data = data, distance = "pscore") %>% 
                 rownames_to_column(var = 'plot_cluster')  %>% 
                   mutate(dataset = 'PSM')

MDM_data2 <- match.data(MDM_match2, data = data) %>% 
                rownames_to_column(var = 'plot_cluster')  %>% 
                   mutate(dataset = 'MDM')
```

```r
all_data <- data %>%  
                rownames_to_column(var = 'plot_cluster') %>% 
                mutate(dataset = 'Unmatched') %>%
                bind_rows(PSM_data, MDM_data, PSM_data2, MDM_data2)
```

```r
all_data_stand <- all_data %>%   
                    select(plot_cluster, dataset, subclass) %>%
                    left_join(data_stand, by = 'plot_cluster')
```

```r
##join the two datasets together
smd_df <- smd %>%
            filter(conf != 'distance')
```

```r
##create 10000 randomly paired datasets for comparrison

randoms <- data.frame()

set.seed(123) 

for (i in 1:1000) {
        j = runif(1, 35, 323) #randomly select the nubmer of matched pairs max is total nubmer of treated min is ~ 50% of treated
        df <- data %>% 
                    rownames_to_column('plot_cluster') %>%
                    group_by(treatment) %>% 
                    slice_sample(n = j) %>%
                    mutate(subclass = seq(1:j)) %>%
                    ungroup() %>% mutate(dataset = paste0('dataset',i))
    
        randoms <- bind_rows(randoms, df)
}
```

```r
## calcualte SMD for the randomly paired datasets

trt_sd_ran <- data %>%
                filter(treatment == 1) %>% 
                    select(all_of(conf)) %>%
                                summarize(across(all_of(conf), ~sd(.x))) %>%
                                pivot_longer(cols = all_of(conf), names_to = 'conf', values_to = 'trt_sd')

smd_ran <- randoms %>%
                select(dataset, treatment, all_of(conf)) %>%
                        group_by(dataset, treatment) %>%

                           summarize(across(all_of(conf), list(mean = mean, var = var),.names = "{.col}-{.fn}")) %>%
                            ungroup() %>%
                             group_by(dataset) %>%
                             arrange(treatment) %>%
                              summarize(across(.cols = ends_with('mean'), ~ (last(.x) - first(.x))), 
                                         across(.cols = ends_with('var'), ~ (last(.x)/first(.x)))) %>%
                              pivot_longer(cols =  !dataset, names_to = c('conf', 'measure'),  names_sep= "-", values_to = 'value') %>%

                              pivot_wider(names_from = 'measure', values_from = 'value') %>%
                              rename('mean_dif' = 'mean', 'var_ratio' = 'var') %>%

                              left_join(trt_sd_ran, by = c('conf')) %>%
                                    mutate(smd = mean_dif / trt_sd, abs_smd = abs(mean_dif) / trt_sd) %>%
                                ungroup()
                                           
```

```r
##select pallette
pal <- as.list(met.brewer('Archambault'))
```

```r
##produce plot of smds for all matches  
options(repr.plot.width=12, repr.plot.height=10)

(smd_plot <- ggplot() + 
            geom_violin(data = smd_ran, aes(conf, abs_smd), fill = '#F8F9F9') +
 
            geom_hline(yintercept = 0, linetype = 'dashed') + 
            geom_hline(yintercept = 0.1, linetype = 'dotted') + 
 
            geom_point(data = filter(smd_df, dataset == 'Unmatched'), aes(conf, abs_smd, colour = dataset), stroke = 2, size = 4) +
            geom_point(data = filter(smd_df, dataset != 'Unmatched'), aes(conf, abs_smd, colour = dataset), stroke = 2, size = 4,
                       position = position_dodge(width = 0.2)) +
            coord_flip() +
            theme_cowplot() +
            scale_colour_manual(values = c(pal[2], pal[1], pal[6], pal[5], pal[4])) +
            labs(y='Absolute standardizered mean difference') +
            theme(text = element_text(size = 18),  
                     axis.title.y =element_blank(), 
                     axis.text=element_text(size = 18),
                     legend.position = c(0.75, 0.9), legend.title = element_blank())
)
```

```r
ggsave(smd_plot, 'png', filename = 'figures/appendixS4.png', width = 12, height = 10)
```

```r
##plot of ratio of varaince for each dataset coampred to randomly generated pairs
options(repr.plot.width=12, repr.plot.height=10)

(r_var_plt <- ggplot() + 
            geom_violin(data = smd_ran, aes(conf, var_ratio), fill = '#F8F9F9') +
 
            geom_hline(yintercept = 0.5) + 
            geom_hline(yintercept = 1, linetype = 'dashed') +
            geom_hline(yintercept = 2, linetype = 'dotted') + 
            geom_point(data = filter(smd_df, dataset == 'unmatched'), aes(conf, var_ratio, colour = dataset), stroke = 2, size = 4) +
            geom_point(data = filter(smd_df, dataset != 'unmatched'), aes(conf, var_ratio, colour = dataset), stroke = 2, size = 4,
                       position = position_dodge(width = 0.2)) +
            coord_flip() +
            theme_cowplot() +
            scale_colour_manual(values = c(pal[2], pal[1], pal[6], pal[5], pal[4])) +
            labs(y='Ratio of variance') +
            theme(text = element_text(size = 18),  
                     axis.title.y =element_blank(), 
                     axis.text=element_text(size = 18),
                     axis.text.y=element_blank(),
                     legend.position = c(0.75, 0.9), legend.title = element_blank())
)
```

```r
##calcualte pairwise difference for each dataset
all_dif <- all_data_stand %>%
                filter(dataset != 'Unmatched') %>%
            group_by(dataset, subclass) %>%
            arrange(treatment) %>%
            summarize(across(.cols = all_of(conf), ~ abs((last(.x) - first(.x))))) %>%  ##this is treatment - control
            pivot_longer(cols = all_of(conf), names_to = 'conf', values_to = 'value') %>%
            ungroup()
```

```r
##calcualte pairwise difference for each of the randomly paired datasets
ran_dif <- randoms %>%
            select(dataset, plot_cluster, subclass) %>%
            left_join(data_stand, by = 'plot_cluster') %>%
            group_by(dataset, subclass) %>%
            arrange(treatment) %>%
            summarize(across(.cols = all_of(conf), ~ abs((last(.x) - first(.x))))) %>%  ##this is treatment - control
            ungroup() %>%
            pivot_longer(cols = all_of(conf), names_to = 'conf', values_to = 'value') %>%
            group_by(dataset, conf) %>%
            summarize(value = mean(value), se = sd(value)/sqrt(length(value))) %>%
            ungroup()
```

```r
all_dif2 <- all_dif %>%
                group_by(dataset, conf) %>%
                  summarize(mean = mean(value), se = sd(value)/sqrt(length(value)))
```

```r
options(repr.plot.width=10, repr.plot.height=10)
##pairwise difference plot for each dataset
(pair_dif_plot <- ggplot() + 
                     geom_violin(data = ran_dif, aes(conf, value), fill = '#F8F9F9') +

                    geom_pointrange(data = all_dif2, 
                                    aes(x = conf, y = mean, ymin = mean -(1.96*se), ymax = mean + (se*1.96), colour = dataset),
                                       position = position_dodge(width =0.3), stroke = 2, size = 1) + 
                    coord_flip() +
                    labs(y='Mean difference between matched pairs') +
                    theme_cowplot() +
                    scale_colour_manual(values = c(pal[2], pal[1], pal[6], pal[5], pal[4])) +
                    scale_shape_manual(values = c('yes'=19, 'no'=15), labels = c('sig. (p < 0.05)', 'N.S.')) +
                   
                    theme(text = element_text(size = 18),  
                     axis.title.y=element_blank(), 
                     axis.text=element_text(size = 18),
                     #axis.text.y=element_blank(),
                     legend.position = 'top', legend.title = element_blank()) +
                     guides(color = 'none')
 )
```

```r
options(repr.plot.width=18, repr.plot.height=10)

dif_grid <-  plot_grid(pair_dif_plot, r_var_plt)
dif_grid
```

```r
ggsave(dif_grid, 'png', filename = 'figures/appendixS6.png', width = 20, height = 10)
```
